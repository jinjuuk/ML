```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
data = pd.read_csv(url)
X = data.iloc[:,:13]
y = data.iloc[:,13]
X
```

## 이상치 제거

```python
columns = list(X.columns)

for column in columns:
    Q1 = np.percentile(X[column],25)
    Q3 = np.percentile(X[column],75)
    IQR = Q3 - Q1
    upper_outliers = Q3 + IQR*1.5
    lower_outliers = Q1 - IQR*1.5

    if column == 'chas':
        pass
    else:
        X[f'scaled_{column}'] = X[column].where((X[column] > lower_outliers) & (X[column] < upper_outliers))
X
```

```python
X = X.drop(columns=columns, axis=1)
X = X.drop(columns=['scaled_zn'])
X
```

```python
X.isna().sum()
```

```python
re_columns = list(X.columns)
for re_column in re_columns:
    X[re_column] = X[re_column].fillna(X.loc[:, re_column].mean())
```

```python
X.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PolynomialFeatures
```

```python
def scale_converter(X=None, degree=None, method='None'):
    if method == 'standard':
        scaled_df = StandardScaler().fit_transform(data)
    elif method == 'minmax':
        scaled_df = MinMaxScaler().fit_transform(data)
    elif method == 'log':
        scaled_df = np.log1p(data)
    else:
        scaled_df = data

    
    
    if degree != None:
        scaled_df = PolynomialFeatures(degree=degree, include_bias=False).fit_transform(scaled_df)
    return scaled_df
        
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
for i in range(1,5):
    X_standard = scale_converter(X,i,'None')
    X_train, X_test, y_train, y_test = train_test_split(X_standard, y, test_size=0.1, random_state=123)
    
    alphas = np.linspace(0.1,100,20)
    ridge_cv = RidgeCV(alphas=alphas, cv=5).fit(X_train, y_train)
    yhat = ridge_cv.predict(X_test)
    
    print(f'order: {i} R^2: {r2_score(y_test, yhat)} for best alpha: {ridge_cv.alpha_}')
    print(f'order: {i} MSE: {mean_squared_error(y_test, yhat)} for best alpha: {ridge_cv.alpha_}')
    print(f'order: {i} MAE: {mean_absolute_error(y_test, yhat)} for best alpha: {ridge_cv.alpha_}')
    print()
```
